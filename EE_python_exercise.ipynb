### بخش دوم

<p>ابتدا تابع رگرسیون خطی را فراخوانی می‌کنیم.</p>

```python
from sklearn.linear_model import LinearRegression
```

<p>سپس متغیرهای مورد نیاز برای فیت کردن مدل را تعریف می‌کنیم.</p>

```python
X_salary = company_data['Years'].values.reshape(-1, 1)
Y_salary = company_data['Salary'].values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(X_salary, Y_salary)
```

<p>بعد از فیت کردن مدل، تابع را بر اساس آن می‌نویسیم.</p>

```python
salary_intercept = model.intercept_[0]
salary_coefficient = model.coef_[0]
```

### خواسته اول

<P> برای محاسبه‌ی سود، نیاز به درآمد و هزینه‌های هر سال داریم بنابراین توابعی  برای محاسبه‌ی این موارد می‌نویسیم.</p>

<p>اولین تابع ما، هزینه‌ی محصول اول را پیشبینی می‌کند. در این تابع ابتدا بررسی می‌کند که آیا دیتای مورد نظر موجود می‌باشد و در صورت موجود نبودن، آن را پیشبینی می‌کند.</p>

```python
def material1_cost_prediction(year, rate=material1_rate):
    
    last_year_available = updated_company_data['Years'].iloc[-1]

    period = year - last_year_available
    last_material1_cost_available = updated_company_data['material 1'].iloc[-1]
    return p_to_f(period, rate, last_material1_cost_available)

```

<p>در این تابع، هزینه‌ی محصول 2 محاسبه می‌شود. به طور کلی این تابع مانند تابع قبلی کار می‌کند و تنها فرق مهمی که دارند، این است که در این تابع از مدل رگرسیون استفاده می‌کنیم و به دلیل نحوه‌ی فیت کردن مدل، نیازی به توابع اقتصاد مهندسی نداریم.</p>

```python
def material2_cost_prediction(year, rate=rate):
    
    
        material2_cost = p_to_f(year-1990, rate, predict_material2(year))
        demand_function = caculate_demand(year)
        demand = demand_prediction(demand_function)
        
        if demand[0] > 60:
            return (material2_cost * 0.82)[0]
        elif 40 < demand[0] <= 60:
            return (material2_cost * 0.9)[0]
        elif 20 < demand[0] <= 40:
            return (material2_cost * 0.95)[0]
        else:
            return material2_cost[0]
```

<p>دقیقا مانند توابع قبلی، توابعی برای محاسبه‌ی حقوق، تقاضا، قیمت و سود تعریف می‌کنیم.</p>

```python
def salary_payable(year):
    

        salary_estimated = predict_salary(year)
        return salary_estimated[0]
        
```

```python
def demand_estimation(year):
    

        demand_function = caculate_demand(year)
        estimated_demand = demand_prediction(demand_function)
        return estimated_demand[0]
```

```python
def price_estimation(year):
    

        price = predict_final_price(year)
        return price[0]
```

```python
def profit(year):
    
    estimated_cost = (material1_cost_prediction(year) + material2_cost_prediction(year))* demand_estimation(year) + salary_payable(year)
    estimated_revenue = demand_estimation(year) * price_estimation(year)
    net_profit = estimated_revenue - estimated_cost
    
    return net_profit
```
